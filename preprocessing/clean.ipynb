```python
import json
```

```python
class UkrainianCitiesNamesComparator(dict):
    ALPHABET = 'АБВГҐДЕЄЖЗИІЇЙКЛМНОПРСТУФХЦЧШЩЬЮЯ'
    
    @staticmethod
    def get_first_different_letters_indices(word1, word2):
        word1 = word1.upper()
        word2 = word2.upper()
        
        i = 0
        while i < min(len(word1), len(word2)):
            first_letter_index = UkrainianCitiesNamesComparator.ALPHABET.index(word1[i])
            second_letter_index = UkrainianCitiesNamesComparator.ALPHABET.index(word2[i])

            if first_letter_index != second_letter_index:
                return first_letter_index, second_letter_index
            i += 1
        return 0, 0

    def __lt__(self, other):
        index1, index2 = UkrainianCitiesNamesComparator.get_first_different_letters_indices(self['title'], other['title'])       
        return index1 < index2

# sort nodes in ukrainian alphabetical order
# nodes = sorted(nodes, key=UkrainianCitiesNamesComparator)
```

```python
with open('nodes.json', encoding='utf-8-sig') as file:
    nodes = json.load(file)

with open('graph.json', encoding='utf-8-sig') as file:
    graph = json.load(file)

# nodes
def get_title_by_id(id):
    for node in nodes:
        if node['id'] == id:
            return node['title']
        
def get_neighbours_by_id(id):
    ns = []
    for rel in graph:
        if rel['start'] == id:
            ns.append(
                {
                    'title': get_title_by_id(rel['end']),
                    'distance': rel['distance']
                })

    return ns

for node in nodes:
    ns = get_neighbours_by_id(node['id'])
    node['neighbours'] = ns
    # del node['id']

for node in nodes:
    del node['id']

nodes
```

```python
with open('nodes.json', 'w', encoding='utf-8-sig') as file:
    json.dump(nodes, fp=file, ensure_ascii=False, indent=4)
```

```python
with open('graph.json', 'w', encoding='utf-8-sig') as file:
    json.dump(graph, fp=file, ensure_ascii=False)
```
